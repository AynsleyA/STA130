1. The difference between simple linear regression and multiple linear regression is the number of independent variables that affect the dependent variable. In simple linear regression there is one independent variable while there can be many in multiple linear regression. Because multiple linear regression includes many variables it is more applicable to real-life situations. It can also be useful to seeing how only one variable affects the dependent variable because the other variables will be considered constant. By having more variables, multiple linear regression will often result in better predictions.
2. One difference between using a continuous variable and an indicator variable in simple linear regression is the type of data that you use them with. A continuous variable would be used for continuous, numerical data while an indicator variable would be used for categorical data. The other difference would be based off of what you are looking for. An indicator variable would be used when there are two groups and you are looking for the difference between them, while continuous variables work better for when you need specific quantitative values. This is the same for multiple linear regression it will just be a bit more complicated.

Simple linear regression with a continuous variable:

$$ Y_i = \beta_0 + \beta_1 x_i + \epsilon_i $$

Simple linear regression with an indicator variable:

$$ Y_i = \beta_0 + \beta_1 D + \epsilon $$

3. When a single indicator variable is used alongside continuous variables the result is two regression lines, seperated by group, that show the effect that the continuous variable has on the dependent variable. The slope will be the same for each line since the relationship between the independent variable and the dependent variable will be the same, the difference between the lines will be found in the intercepts which will model the difference between the actual values of the two groups. Since this takes two variables this can only happen with multiple linear regression and not simple linear regression.

Simple linear regression:

$$ Y_i = \beta_0 + \beta_1 x_i + \epsilon_i $$

Multiple linear regression:

$$ Y_i = \beta_0 + \beta_1 x_i + \beta_2 D + \epsilon_i $$

4. When an interaction between the continuous variable and the indicator variable is added this allows for the slope between the two groups to be different. This makes it easier to get more insight into the data since it isn't very often that the slopes of two different groups are going to be the same so now we can do better comparision.

Multiple linear regression:

$$ Y_i = \beta_0 + \beta_1 x_i + \beta_2 D + \beta_3 (x_i  D) + \epsilon_i $$

5. A multiple linear regression model based only on indicator variables derived from a non-binary categorical variable will have one less indicator variable than categories and will use all these indicator variables to look at the differences between the groups. This will happen by using something similar to if-else method. For example if we have three indicator variables, we would have four groups. The three indicator variables will each be assigned to a group but if all of those indicator variables result in 0 we know that the proper group for the variable is the fourth group. This is how even with non-binary categories the method to sort them is still binary. This will give 4 regression lines where the only difference is in the intercept because with only indicator variables not much else can be done. 

Multiple linear regression with only indicator variables:

$$ Y_i = \beta_0 + \beta_1 D_1 + \beta_2 D_2 + ... + \beta_k D_k + \epsilon \quad \text{where k is the number of indicator variables} $$


Summary: Simple vs. Multiple Linear Regression:

Simple Linear Regression models the relationship between a dependent variable and one independent variable.
Multiple Linear Regression extends this to include multiple independent variables, allowing for the analysis of combined effects.
Benefits of Multiple Linear Regression:

Captures relationships between the dependent variable and multiple predictors.
Accounts for confounding variables and measures marginal effects.
Improves predictive accuracy by incorporating more information.
Continuous vs. Indicator Variables:

Continuous variables represent quantitative relationships, with coefficients indicating the effect of one-unit changes.
Indicator (dummy) variables represent categorical data, showing differences in the dependent variable between groups.
Adding an Interaction Term:

An interaction between a continuous and an indicator variable allows the slope of the continuous variableâ€™s effect to differ between groups. This creates separate, non-parallel regression lines for each group.
Modeling with Only Indicator Variables:

A model with only indicator variables derived from a categorical variable compares group means. It captures category-specific differences in the dependent variable but assumes no variation within categories.

```python
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

# Load data once
data = pd.read_csv("CSCS_data_anon.csv")

# Check for missing values and handle them properly (e.g., fill with 0 or drop NaNs if appropriate)
data = data.fillna(0)  # Or use other imputation methods, as needed

# Convert categorical variables to binary (0/1)
data['children'] = (data['GEO_housing_live_with_children'] > 0).astype(int)
data['partner'] = (data['GEO_housing_live_with_partner'] > 0).astype(int)

# Ensure the dependent variable is numeric
data['WELLNESS_life_satisfaction'] = pd.to_numeric(data['WELLNESS_life_satisfaction'], errors='coerce')

# Handle missing or invalid entries in the dependent variable (e.g., drop rows with NaNs in the target)
data = data.dropna(subset=['WELLNESS_life_satisfaction'])

# Define the formula for the logistic regression model
linear_model_specification_formula = 'WELLNESS_life_satisfaction ~ children + partner'

# Fit the logistic regression model
log_reg_fit = smf.ols(linear_model_specification_formula, data=data).fit()

# Display the model summary
print(log_reg_fit.summary())

# Boxplot for 'children' vs 'WELLNESS_life_satisfaction'
plt.figure(figsize=(10, 6))
sns.boxplot(x='children', y='WELLNESS_life_satisfaction', data=data, palette="Set2")
plt.title("Life Satisfaction by Living with Children")
plt.xlabel("Living with Children (0 = No, 1 = Yes)")
plt.ylabel("Life Satisfaction")
plt.show()

# Boxplot for 'partner' vs 'WELLNESS_life_satisfaction'
plt.figure(figsize=(10, 6))
sns.boxplot(x='partner', y='WELLNESS_life_satisfaction', data=data, palette="Set3")
plt.title("Life Satisfaction by Living with Partner")
plt.xlabel("Living with Partner (0 = No, 1 = Yes)")
plt.ylabel("Life Satisfaction")
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
import plotly.graph_objects as go

# Load the diabetes dataset from sklearn
data = load_diabetes()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')

# Combine X and y into one DataFrame for easier handling
df = pd.concat([X, y], axis=1)

# Define the formula for the model (for example, predicting 'target' using 'bmi' and 'age' as features)
linear_form = 'target ~ bmi + age'

# Number of repetitions
reps = 100

# Arrays to store the RÂ² values for each iteration
in_sample_Rsquared = np.array([0.0] * reps)
out_of_sample_Rsquared = np.array([0.0] * reps)

for i in range(reps):
    # Randomly split the data into training and testing sets (train size 70%, test size 30%)
    df_train, df_test = train_test_split(df, train_size=0.7)
    
    # Fit the model on the training data
    final_model_fit = smf.ols(formula=linear_form, data=df_train).fit()
    
    # Calculate in-sample RÂ² (RÂ² on the training data)
    in_sample_Rsquared[i] = final_model_fit.rsquared
    
    # Calculate out-of-sample RÂ² (RÂ² on the testing data)
    out_of_sample_Rsquared[i] = np.corrcoef(df_test.target, final_model_fit.predict(df_test))[0, 1] ** 2

# Create a DataFrame to store the results
result_df = pd.DataFrame({
    "In Sample Performance (Rsquared)": in_sample_Rsquared,
    "Out of Sample Performance (Rsquared)": out_of_sample_Rsquared
})

# Create a scatter plot with Plotly
fig = px.scatter(result_df, x="In Sample Performance (Rsquared)", y="Out of Sample Performance (Rsquared)")

# Add a reference line (y=x) to visualize ideal performance
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], name="y=x", line_shape='linear'))

# Show the plot
fig.show(renderer='png')
```

This code and visualization shows how a model works on the sample it was tested on, in sample performance, and how accurate it is for samples other than the one it was tested on, out of sample perofrmance. For a model that is accurate in and out of sample the R squared would be high and the points would be on the refernce line. In this case we've got our points lower on the x and y axis so the model isn't super accurate, sometimes it overfits and sometimes it underfits. This test can be used to check performance of a specific model.

Summary: The Task: You want to run a model many times, each time splitting the data randomly into training and testing sets. For each split, you calculate:
In-sample RÂ²: How well the model fits the training data.
Out-of-sample RÂ²: How well the model works on new, unseen data (testing data).
What We Did:
We used the diabetes dataset to build a linear regression model.
We ran the model 100 times with different random splits of the data.
We calculated and stored both the in-sample and out-of-sample RÂ² values each time.
Purpose: The goal was to check if the model is overfitting (doing well on training data but poorly on test data) or underfitting (doing poorly on both).
Result: We visualized the RÂ² values to see if the model performs well and consistently on both training and testing data.
